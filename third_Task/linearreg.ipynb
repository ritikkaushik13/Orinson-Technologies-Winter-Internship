```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipeline = Pipeline([
    ('poly',PolynomialFeatures()),
    ('model',LinearRegression())
])
```

```python
pipeline
```

```python
grid_params = {
    'poly__degree':[2,3,4,5],
    'model__fit_intercept':[True,False]
}
```

```python
gridss = GridSearchCV(pipeline,grid_params,scoring='r2',cv=5)
```

```python
gridss
```

```python
gridss.fit(X_train,y_train)
```

```python
gridss.best_estimator_
```

```python
gridss.best_params_
```

```python
gridss.best_score_
```

```python
best_model = gridss.best_estimator_
y_pred = best_model.predict(X_test)

# Evaluate on the same data (example purposes)
r2 = r2_score(y_test, y_pred)
print(f"R^2 on full data: {r2}")
```
